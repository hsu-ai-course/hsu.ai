```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics

data = []
for line in open("../datasets/ml/clustering.20.txt", 'r'):
    x, y = line.strip().split('   ')
    data.append([int(x), int(y)])
X = np.array(data)

plt.figure(figsize=(10,10))
plt.plot(X[:,1], X[:,0], 'r.')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering, DBSCAN, KMeans

kmeans = KMeans(n_clusters=20,)
aclust = AgglomerativeClustering(n_clusters=20,)
dbscan = DBSCAN(
        eps=600,             # max distance to consider neighbours
        min_samples=15,      # neighbours for core point
        metric='euclidean'   # what is distance?
    )

km = kmeans.fit(X)
ac = aclust.fit(X)
db = dbscan.fit(X)
```

```python
# DBSCAN SPECIFIC
# Number of clusters in labels, ignoring noise if present.
# -1 stands for noise
n_clusters_ = len(set(db.labels_)) - (1 if -1 in db.labels_ else 0)
n_noise_ = list(db.labels_).count(-1)

print(f'Estimated number of clusters: {n_clusters_}')
print(f'Estimated number of noise points {n_noise_}')
```
