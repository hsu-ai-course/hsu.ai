```python
import numpy as np
i = np.linspace(0, 1, 11)
j = np.linspace(0, 1, 11)
i, j = np.meshgrid(i, j)
X = np.array(list(zip(i.flatten(), j.flatten())))
```

```python
def unknown(point):
    if point[0] == point[1] and point[0] < 0.3:
        return 0
    return int(abs(point[0] ** 2 - point[1] ** 3) < 0.6)


def unknown2(point):
    return int((point[0] - point[1]) * 10) % 2 

## TODO anothe propose a function
```

```python
y = np.array([unknown(p) for p in X])
```

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(hidden_layer_sizes=(10, 10, 10, 10, 10, 10), random_state=1, max_iter=4000).fit(X, y)
```

```python
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt

# Plotting decision regions
plot_decision_regions(X, y, clf=clf, legend=2)

# Adding axes annotations
plt.xlabel('Some data')
plt.title('ANN for unknows function learning')

plt.show()
```
