```python
# detect the field and all checkers with their positions (image coordinates and board coordinates)
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.cvtColor(cv2.imread('datasets/images/checkers.jpg'), cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
plt.imshow(gray, cmap="gray")
plt.show()
plt.imshow(hsv[:,:,0], cmap="gray")
plt.show()
plt.imshow(hsv[:,:,1], cmap="gray")
plt.show()
```
