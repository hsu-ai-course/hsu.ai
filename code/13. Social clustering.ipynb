## Read the data from a social graph
Originally from [here](https://snap.stanford.edu/data/email-Eu-core.html).

```python
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns


data_path = 'datasets/ml/facebook_combined.txt'
G_social = nx.read_edgelist(data_path)
```

## Take a look on the graph

NB Layout is a computationally intense operation, though it can take 1-2 minutes.

```python
pos = nx.spring_layout(G_social, k=0.1)
plt.rcParams.update({'figure.figsize': (15, 10)})
nx.draw_networkx(
    G_social, 
    pos=pos, 
    node_size=0, 
    edge_color="#444444", 
    alpha=0.05, 
    with_labels=False)
```
